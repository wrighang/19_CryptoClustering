```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
market_data_df = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
market_data_df.head(10)
```

```python
# Generate summary statistics
market_data_df.describe()
```

```python
#check data types
market_data_df.info()
```

```python
# Plot your data to see what's in your DataFrame
market_data_df.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
#display dimensions of DataFrame (rows, columns)
market_data_df.shape
```

```python
#display column names
market_data_df.columns
```

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(market_data_df[["price_change_percentage_24h",	"price_change_percentage_7d",	"price_change_percentage_14d",
                                                                    "price_change_percentage_30d",	"price_change_percentage_60d",	"price_change_percentage_200d",	"price_change_percentage_1y"]])
```

```python
#diplay first three rows of scaled data
market_data_scaled[0:3]
```

```python
# Create a DataFrame with the scaled data
market_data_scaled_df = pd.DataFrame(market_data_scaled, columns = ["price_change_percentage_24h",	"price_change_percentage_7d",	"price_change_percentage_14d",
                                                                    "price_change_percentage_30d",	"price_change_percentage_60d",	"price_change_percentage_200d",	"price_change_percentage_1y"])

# Copy the crypto names from the original DataFrame
market_data_scaled_df["coin_id"] = market_data_df.index

# Set the coin_id column as index
market_data_scaled_df = market_data_scaled_df.set_index("coin_id")

# Display the scaled DataFrame
market_data_scaled_df.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `market_data_scaled_df`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    model = KMeans(n_clusters=i, random_state=1)
    model.fit(market_data_scaled_df)
    inertia.append(model.inertia_)
model
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)
elbow_df.head()
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow_curve_plot = elbow_df.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
elbow_curve_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for `k` is the elbow point where `k` = 4. This is the point where adding more clusters does not significantly reduce inertia. 

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=42)
```

```python
# Fit the K-Means model using the scaled DataFrame
model.fit(market_data_scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
crypto_clusters = model.predict(market_data_scaled_df)

# Print the resulting array of cluster values.
print(crypto_clusters)
```

```python
# Create a copy of the scaled DataFrame
market_data_scaled_predict_df = market_data_scaled_df.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
market_data_scaled_predict_df["crypto_clusters"] = crypto_clusters

#sortying by crytpo_clster b/c the legend was orignally not in order was showing 2, 0, 3, 1
market_data_scaled_predict_df = market_data_scaled_predict_df.sort_values(by="crypto_clusters")

# Display the copy of the scaled DataFrame
market_data_scaled_predict_df.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
crypto_cluster_scatter_plot = market_data_scaled_predict_df.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="crypto_clusters",
    hover_cols=["coin_id"],
)

crypto_cluster_scatter_plot
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
market_data_pca = pca.fit_transform(market_data_scaled_df)

# View the scaled PCA data
market_data_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component. 
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 37.2% of the variance can be explained by the 1st principal component, the 2nd princiipal component is 34.7% and the 3rd principal component is 17.6%. The total explained variance is about 89.5%. 

```python
# Create a new DataFrame with the PCA data.
market_data_pca_df = pd.DataFrame(
    market_data_pca, 
    columns=["PC1", "PC2", "PC3"]
)

# Copy the crypto names from the original scaled DataFrame
market_data_pca_df["coin_id"] = market_data_df.index

# Set the coin_id column as index
market_data_pca_df = market_data_pca_df.set_index("coin_id")

# Display the scaled PCA DataFrame
market_data_pca_df.head()
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11)) 
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `market_data_pca_df`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    pca_model = KMeans(n_clusters=i, random_state=1)
    pca_model.fit(market_data_pca_df)
    inertia.append(pca_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data_pca = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_pca_df = pd.DataFrame(elbow_data_pca)
elbow_pca_df.head()
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
pca_elbow_curve_plot = elbow_pca_df.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
    )
pca_elbow_curve_plot
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** When using the PCA data, the best value for `k` is the elbow point where `k` = 4. This is the point where adding more clusters does not significantly reduce inertia. 


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No, it does not differ. Both elbow curves show `k` = 4. 

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the PCA data
model.fit(market_data_pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
crypto_clusters_pca = model.predict(market_data_pca_df)

# Print the resulting array of cluster values.
print(crypto_clusters_pca)
```

```python
# Create a copy of the scaled PCA DataFrame
pca_predictions_df = market_data_pca_df.copy()

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
pca_predictions_df["stock_cluster"] = crypto_clusters_pca

#sortying by PCA_prediction b/c the legend was orignally not in order was showing 2, 0, 3, 1
pca_predictions_df = pca_predictions_df.sort_values(by="stock_cluster")

# Display the copy of the scaled PCA DataFrame
pca_predictions_df.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
pca_scatter_plot = pca_predictions_df.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="stock_cluster",
    hover_cols = ("coin_id")
)
pca_scatter_plot
```
